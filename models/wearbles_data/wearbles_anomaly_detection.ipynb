```python
# =========================
# 1. Imports
# =========================
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import shuffle
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, Dropout


# =========================
# 2. Load Data
# =========================
activity = pd.read_csv("/content/data/activity_environment_data.csv")
health = pd.read_csv("/content/data/personal_health_data.csv")
digital = pd.read_csv("/content/data/digital_interaction_data.csv")

activity.columns = activity.columns.str.lower()
health.columns = health.columns.str.lower()
digital.columns = digital.columns.str.lower()


# =========================
# 3. Merge + Clean
# =========================
data = pd.merge(activity, health, on=['user_id', 'timestamp'])
data = pd.merge(data, digital, on=['user_id', 'timestamp'])

data['exercise_type'] = data['exercise_type'].fillna("None")
data['exercise_intensity'] = data['exercise_intensity'].fillna("None")
data['medical_conditions'] = data['medical_conditions'].fillna("None")
data['alcohol_consumption'] = data['alcohol_consumption'].fillna("None")

data['timestamp'] = pd.to_datetime(data['timestamp'])
data = data.sort_values(by=['timestamp'])


# =========================
# 4. Feature Selection
# =========================
features = [
    'heart_rate',
    'steps',
    'calories_burned'
]

labels = data['anomaly_flag'].values


# =========================
# 5. Normalize
# =========================
scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])


# =========================
# 6. Create Sequences
# =========================
sequence_length = 50
X, y = [], []

values = data[features].values

for i in range(len(values) - sequence_length):
    X.append(values[i:i+sequence_length])
    y.append(labels[i+sequence_length])

X = np.array(X)
y = np.array(y)


# =========================
# 7. Train/Test Split
# =========================
normal_data = X[y == 0]
anomaly_data = X[y == 1]

X_train = normal_data

anomaly_sample = anomaly_data[:int(0.2 * len(normal_data))]

X_test = np.concatenate([normal_data, anomaly_sample])
y_test = np.concatenate([
    np.zeros(len(normal_data)),
    np.ones(len(anomaly_sample))
])

X_test, y_test = shuffle(X_test, y_test)


# =========================
# 8. Model (TCN-like Autoencoder)
# =========================
timesteps = X.shape[1]
num_features = X.shape[2]

input_layer = Input(shape=(timesteps, num_features))

x = Conv1D(64, 3, padding='same', activation='relu')(input_layer)
x = Dropout(0.2)(x)
x = Conv1D(32, 3, padding='same', activation='relu')(x)

x = Conv1D(32, 3, padding='same', activation='relu')(x)
x = Dropout(0.2)(x)
output_layer = Conv1D(num_features, 3, padding='same')(x)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer='adam', loss='mse')


# =========================
# 9. Train
# =========================
model.fit(
    X_train, X_train,
    epochs=30,
    batch_size=32,
    validation_split=0.1
)


# =========================
# 10. Reconstruction Error
# =========================
train_pred = model.predict(X_train)
train_mse = np.mean(np.square(X_train - train_pred), axis=(1, 2))

threshold = np.percentile(train_mse, 80)


X_pred = model.predict(X_test)
mse = np.mean(np.square(X_test - X_pred), axis=(1, 2))

ae_pred = (mse > threshold).astype(int)


# =========================
# 11. Rule-Based Detection
# =========================
# Recreate rule features from last timestep of each sequence
last_steps = X_test[:, -1, 1]
last_hr = X_test[:, -1, 0]
last_cal = X_test[:, -1, 2]

rule_pred = (
    (last_hr > 1.5) |     # high normalized HR
    (last_steps > 2.0) |  # unusually high steps
    (last_cal > 2.0)      # high calories
).astype(int)


# =========================
# 12. Hybrid Prediction
# =========================
final_pred = (ae_pred | rule_pred).astype(int)


# =========================
# 13. Evaluation
# =========================
print("Confusion Matrix:")
print(confusion_matrix(y_test, final_pred))

print("\nClassification Report:")
print(classification_report(y_test, final_pred))


# =========================
# 14. Save Model
# =========================
#model.save("/content/drive/MyDrive/final_model.h5")

#import joblib
#joblib.dump(scaler, "/content/drive/MyDrive/final_scaler.pkl")
```

```python
import pandas as pd

# Convert timestamp
steps['ActivityHour'] = pd.to_datetime(steps['ActivityHour'])
calories['ActivityHour'] = pd.to_datetime(calories['ActivityHour'])
intensity['ActivityHour'] = pd.to_datetime(intensity['ActivityHour'])
hr_hourly['ActivityHour']=pd.to_datetime(hr_hourly['ActivityHour'])
# Merge datasets
data = steps.merge(calories, on=['Id', 'ActivityHour'])
data = data.merge(intensity, on=['Id', 'ActivityHour'])
data=data.merge(hr_hourly,on=['Id','ActivityHour'])

# Select features
data = data[['Id', 'ActivityHour', 'steps', 'calories', 'intensity','heart_rate']]

# Sort
data = data.sort_values(by=['Id', 'ActivityHour'])

print(data.shape)
print(data.head())
```

```python
from sklearn.preprocessing import StandardScaler

features = ['steps', 'calories', 'intensity','heart_rate']

scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])
```

```python
import numpy as np

sequence_length = 24  # 1 day window

X = []

for user in data['Id'].unique():
    user_data = data[data['Id'] == user]
    values = user_data[features].values

    for i in range(len(values) - sequence_length):
        X.append(values[i:i+sequence_length])

X = np.array(X)

print("X shape:", X.shape)
```

```python
X_anomaly = X.copy()

# Inject anomalies in 20% data
num_anomalies = int(0.2 * len(X))

indices = np.random.choice(len(X), num_anomalies, replace=False)

for idx in indices:
    # Random anomaly type
    if np.random.rand() < 0.5:
        # Spike in steps
        X_anomaly[idx, :, 0] *= np.random.uniform(2, 3)
    else:
        # No movement but calories high
        X_anomaly[idx, :, 0] *= 0.1
        X_anomaly[idx, :, 1] *= 2

# Labels
y = np.zeros(len(X))
y[indices] = 1
```

```python
from sklearn.model_selection import train_test_split

# Train only on normal
X_train = X[y == 0]

# Test on mixed
X_test = X_anomaly
y_test = y
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv1D, Dropout
from tensorflow.keras.models import Model

input_layer = Input(shape=(X.shape[1], X.shape[2]))

x = Conv1D(128, 3, padding='same', activation='relu')(input_layer)
x = Dropout(0.3)(x)

x = Conv1D(64, 3, padding='same', activation='relu')(x)
x = Dropout(0.3)(x)

x = Conv1D(32, 3, padding='same', activation='relu')(x)

output_layer = Conv1D(X.shape[2], 3, padding='same')(x)

model = Model(input_layer, output_layer)
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(
    X_train, X_train,
    epochs=20,
    batch_size=32,
    validation_split=0.1
)
```

```python
# Reconstruction error
train_pred = model.predict(X_train)
train_mse = np.mean((X_train - train_pred)**2, axis=(1,2))

threshold = np.mean(train_mse) + 1.2 * np.std(train_mse)

# Test
test_pred = model.predict(X_test)
test_mse = np.mean((X_test - test_pred)**2, axis=(1,2))

y_pred = (test_mse > threshold).astype(int)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
